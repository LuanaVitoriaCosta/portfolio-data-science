```python
# Solicita as três notas dos estudantes
nota1 = float(input('Digite a primeira nota: '))
nota2 = float(input('Digite a segunda nota: '))
nota3 = float(input('Digite a terceira nota: '))

# Calcula a média das três notas
media = (nota1 + nota2 + nota3) / 3

# Imprime a média com duas casas decimais
print(f'A média das notas é {media:.2f}')
```

```python
# Define os valores e seus respectivos pesos
valor1 = 5
peso1 = 1

valor2 = 12
peso2 = 2

valor3 = 20
peso3 = 3

valor4 = 15
peso4 = 4

# Calcula a soma ponderada dos valores
soma_ponderada = (valor1 * peso1) + (valor2 * peso2) + (valor3 * peso3) + (valor4 * peso4)

# Calcula a soma dos pesos
soma_pesos = peso1 + peso2 + peso3 + peso4

# Calcula a média ponderada
media_ponderada = soma_ponderada / soma_pesos

# Exibe o resultado
print(f'A média ponderada dos valores é {media_ponderada:.2f}')
```
